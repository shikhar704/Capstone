{ Sales Outlier detection }
This is achieved by using Inter quartile Range
 

Function for outlier detection, 
calculate Q1, Q3 then IQR. 
Finally calculate upper_range and lower_range

```python
import numpy as np 
def outlier_detection(datacolumn):
 sorted(datacolumn)
 Q1,Q3 = np.percentile(datacolumn , [25,75])
 IQR = Q3 - Q1
 lower_range = Q1 - (1.5 * IQR)
 upper_range = Q3 + (1.5 * IQR)
 return lower_range,upper_range
```

Storing these values

```python
Lower_Bound_Sales,Upper_Bound_Sales = outlier_detection(data1.Sales)
Upper_Bound_Sales
```

#### Chi Square:
The Chi-Square test of independence is a statistical test to determine if there is a significant relationship between 2 categorical variables.  In simple words, the Chi-Square statistic will test whether there is a significant difference in the observed vs the expected frequencies of both variables. 


There are a lot of ways to apply Chi square test, for our purpose we are checking if 2 categorical features are independent or not by stating a Null hypothesis (h0)  and an ALternative hypothesis (h1), the alpha is set to 0.05.

h0 -> The categorical features are independent of each other

h1 -> The categorical features are dependent on each other

We are calculating the chi square statistics value and then comparing it with critical value

This can also be done using the p-value method.

Ship Mode and Segment:

```python
dataset_table=pd.crosstab(data_with_region_code['Ship_Mode'],data_with_region_code['Segment'])
dataset_table
```

```python
dataset_table.info()
dataset_table.values 
```

```python
Observed_Values = dataset_table.values 
print("Observed Values :-\n",Observed_Values)
val=stats.chi2_contingency(dataset_table)
val
```

```python
Expected_Values=val[3]
Expected_Values
```

```python
no_of_rows=len(dataset_table.iloc[0:5,0])
no_of_columns=len(dataset_table.iloc[0,0:5])
ddof=(no_of_rows-1)*(no_of_columns-1)
print("Degree of Freedom:-",ddof)
alpha = 0.05
```

```python
from scipy.stats import chi2
chi_square=sum([(o-e)**2./e for o,e in zip(Observed_Values,Expected_Values)])
chi_square_statistic=chi_square[0]+chi_square[1]
```

```python
print("chi-square statistic:",chi_square_statistic)
critical_value=chi2.ppf(q=1-alpha,df=ddof)
print('critical_value:',critical_value)
if chi_square_statistic>=critical_value:
    print("Accept H1,The 2 categorical variables are not independent of each other")
else:
    print("Accept H0,The 2 categorical variables are independent of each other")
```

SARIMA Model

```python
from datetime import datetime
from statsmodels.tsa.statespace.sarimax import SARIMAX
from datetime import timedelta
train_end = datetime(2014,7,1)
test_end = datetime(2014,12,1)

train_data = data_complete_without_category_M[:train_end]
test_data = data_complete_without_category_M[train_end + timedelta(days=1):test_end]
```

```python
my_order = (5,0,1)
my_seasonal_order = (1, 1, 1, 15)

# define model
model = SARIMAX(train_data, order=my_order, seasonal_order=my_seasonal_order)
```

```python
#fit the model
start = time()
model_fit = model.fit()
end = time()
print('Model Fitting Time:', end - start)
```

```python
#summary of the model
print(model_fit.summary())
```

L2 and L3

```python
#get the predictions and residuals
predictions = model_fit.forecast(len(test_data))
predictions = pd.Series(predictions, index=test_data.index)
residuals = test_data - predictions
```

```python
plt.figure(figsize=(10,4))
plt.plot(residuals)
plt.axhline(0, linestyle='--', color='k')
plt.title('Residuals from SARIMA Model', fontsize=20)
plt.ylabel('Error', fontsize=16)
```

```python
start_date = datetime(2014,1,1)
end_date = datetime(2014,12,1)
plt.figure(figsize=(10,4))

plt.plot(data_complete_without_category_M)
plt.plot(predictions)

plt.legend(('Data', 'Predictions'), fontsize=16)

plt.title('Sales', fontsize=20)
plt.ylabel('Production', fontsize=16)
for year in range(start_date.year,end_date.year):
    plt.axvline(pd.to_datetime(str(year)+'-01-01'), color='k', linestyle='--', alpha=0.2)
```

```python
print('Mean Absolute Percent Error:', round(np.mean(abs(residuals/test_data)),4))
```

```python
print('Root Mean Squared Error:', np.sqrt(np.mean(residuals**2)))
```

Rolling forecast

```python
rolling_predictions = test_data.copy()
for train_end in test_data.index:
    train_data = data_complete_without_category_M[:train_end-timedelta(days=1)]
    model = SARIMAX(train_data, order=my_order, seasonal_order=my_seasonal_order)
    model_fit = model.fit()

    pred = model_fit.forecast()
    rolling_predictions[train_end] = pred
```

```python
rolling_residuals = test_data - rolling_predictions
```

```python
plt.figure(figsize=(10,4))
plt.plot(rolling_residuals)
plt.axhline(0, linestyle='--', color='k')
plt.title('Rolling Forecast Residuals from SARIMA Model', fontsize=20)
plt.ylabel('Error', fontsize=16)
```

```python
plt.figure(figsize=(10,4))

plt.plot(data_complete_without_category_M)
plt.plot(rolling_predictions)

plt.legend(('Data', 'Predictions'), fontsize=16)

plt.title('Sales', fontsize=20)
plt.ylabel('Production', fontsize=16)
for year in range(start_date.year,end_date.year):
    plt.axvline(pd.to_datetime(str(year)+'-01-01'), color='k', linestyle='--', alpha=0.2)
```

```python
print('Mean Absolute Percent Error:', round(np.mean(abs(rolling_residuals/test_data)),4))
```

```python
print('Root Mean Squared Error:', np.sqrt(np.mean(rolling_residuals**2)))
```
